```python
import cv2
import numpy as np
from keras.models import load_model
import matplotlib.pyplot as plt

```

```python
Labels=['Late mild cognitive impairment','Alzheimer_s disease','Cognitively normal','Early mild cognitive impairment']
```

```python
model = load_model('C:\\Users\\Krishang Virmani\\Desktop\\Python CB\\ADNI\\CNN_model.h5')  
def predict_single_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.resize(image,(176,176))
    plt.figure(figsize=(6,4.5))
    plt.subplot(1,1,1)
    plt.imshow(image)
    image = image.reshape(1,176,176,3)
    predictions = model.predict(image)
    predicted_class_index = np.argmax(predictions)
    plt.title(Labels[predicted_class_index],fontdict={"fontsize":25},loc="center")
    plt.show()

```

```python
image_path = "C:\\Users\\Krishang Virmani\\Desktop\\Python CB\\ADNI\\Alzheimer_s Dataset\\test\\NonDemented\\26 (88).jpg"
```

```python
predict_single_image(image_path)
```
